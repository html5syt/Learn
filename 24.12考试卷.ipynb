```python
import csv

i = 1
p = 1
s = 0
data = [["次数(i)", "s", "p"]]

while i <= 30:
    temp = []
    print("次数:", i)
    temp.append(i)
    s = s + p
    print("s:", s)
    temp.append(s)
    p = i
    print("p:", p)
    temp.append(p)
    i = i + 1
    data.append(temp)
print(s)
# 写入 CSV 文件
with open("result.csv", "w", newline="", encoding="ANSI") as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
